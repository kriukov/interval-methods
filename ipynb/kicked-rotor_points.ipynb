```julia
function plot_points(N, k)
    data = zeros(N, 2)
    x0 = [2pi*rand(), 2pi*rand()]

    for i in 1:N
        output = T0(x0, k, i)
        data[i, 1:2] = output
    end
    
    data
end
```
