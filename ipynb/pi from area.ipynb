$y = x^2$

```julia
using PyPlot
```

```julia
x = 0:0.01:1
y = sqrt(1 - x.^2);

plot(x, y)
axis("image")
```

```julia
using PyCall
@pyimport matplotlib.patches as patches
rectangle = patches.Rectangle
```

```julia
function draw_rectangle(x, y, xwidth, ywidth, color="grey")
    ax = gca()
    ax[:add_patch](rectangle((x, y), xwidth, ywidth, facecolor=color, alpha=0.5))
end
    
```

```julia
draw_rectangle(0.1, 0.1, 0.1, 0.1)
axis("image")
```

```julia
step = 0.1
for i in 0.0:step:1.0
    for j in 0.0:step:1.0
        if i^2 + j^2 < 1.0
            color = "red"
        else
            color = "grey"
        end
        
        draw_rectangle(i, j, step, step, color)
    end
end
plot(x, y)
axis("image")
            
            
```

```julia
step = 0.01
for i in 0.0:step:1.0
    for j in 0.0:step:1.0
        if i^2 + j^2 < 1.0
            color = "red"
        else
            color = "grey"
        end
        
        draw_rectangle(i, j, step, step, color)
    end
end

plot(x, y)
axis("image")
            
            
```

```julia
step = 0.001
area = 0.0
for i in 0.0:step:1.0
    for j in 0.0:step:1.0
        
        if i^2 + j^2 < 1.0
            
            area += step^2
            
        end
    end
end
```
