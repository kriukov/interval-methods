```julia
using KrawczykMethod
```

```julia
compose(f::Function, g::Function) = x -> f(g(x))
```

```julia
T1(x) = Interval(r)*(Interval(1/4)-(x-Interval(1/2))^2)
```

```julia
function T(x, p)
	F = T1
	for i = 1:p-1
		F = compose(F, T1)
	end
	return F(x)
end
```

```julia
f(x, p) = T(x, p) - x
```

```julia
r = 3.75
```

```julia
g(x) = f(x, 8)
```

```julia
krawczyk(g, Interval(0, 1), 64)
```

```julia
g(x) = f(x, 10)
```

```julia
krawczyk(g, Interval(0, 1), 64)
```

```julia
krawczyk(g, Interval(0, 1), 128)
```
